```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
data = pd.read_csv("C:\\Users\\ADMIN\\Downloads\\myexcel - myexcel.csv.csv")
data
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
x = data['Salary'].mean()
data['Salary'] = data['Salary'].fillna(data['Salary'].mean())
data
```

```python
data["College"] = data["College"].fillna("Unknown")
data
```

```python
data.isnull().sum()
```

```python
data['Height'] = np.random.uniform(150, 180, size=len(data))
data.head(50)
```

```python
data['Team'].value_counts()
```

```python
data['Team'].value_counts()/len(data)*100
```

```python
grouped = data.groupby('Position')
from IPython.display import display, HTML
for position, group in grouped:
    print(f"Group for position {position}:")
    display(HTML(group.to_html()))
    print("\n")
```

```python
data['Age Group'] = data['Age'].apply(
    lambda age: '20-25' if 20 <= age <= 25 
    else '26-30' if 26 <= age <= 30 
    else '31-35' if 31 <= age <= 35 
    else '36 and above'
)
data
```

```python
age_group_counts = data['Age Group'].value_counts()
print("Number of employees in each age group:\n")
print(age_group_counts)
most_employees_age_group = age_group_counts.idxmax()
print("\nAge group with the most employees:", most_employees_age_group, "with", age_group_counts.max(), "employees.")
```

```python
spending_salary = data.groupby(['Team','Position'])['Salary'].sum()
spending_salary.idxmax()
```

```python
correlation = data['Salary'].corr(data['Age'])
print("THE CORRELATION B/w Salary AND Age IS:",correlation)
```

```python
sns.scatterplot(x="Age" ,y= "Salary",data= data)
plt.ylabel("Salary")
plt.xlabel("Age")
plt.title("correlation b/w Salary and Age")
plt.show()
```
